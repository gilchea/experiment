```python
import json
import numpy as np
import matplotlib.pyplot as plt

# =========================
# LOAD DATA
# =========================

import json

file_path = r"D:\Khóa luận tốt nghiệp\experiment\results\mnist_results.json"

with open(file_path, "r", encoding="utf-8") as f:
    data = json.load(f)

algorithms = ["svrg", "sgd_const", "sgd_best", "sdca", "sag"]

# =========================
# CREATE FIGURE
# =========================

fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# =========================
# 1. LOSS RESIDUAL
# =========================

ax = axes[0, 0]

for algo in algorithms:
    passes = data[algo]["passes"]
    loss = data[algo]["loss_residual"]

    ax.plot(passes, loss, linewidth=2, label=algo.upper())

ax.set_title("Loss Residual vs Effective Passes")
ax.set_xlabel("Effective Passes")
ax.set_ylabel("P(w) - P(w*)")
ax.grid(True)
ax.legend()

# =========================
# 2. TEST ERROR
# =========================

ax = axes[0, 1]

for algo in algorithms:
    passes = data[algo]["passes"]
    test_error = data[algo]["test_error"]

    ax.plot(passes, test_error, linewidth=2, label=algo.upper())

ax.set_title("Test Error vs Effective Passes")
ax.set_xlabel("Effective Passes")
ax.set_ylabel("Test Error (%)")
ax.grid(True)
ax.legend()

# =========================
# 3. SVRG GRADIENT VARIANCE
# =========================

ax = axes[1, 0]

passes = data["svrg"]["passes"]
grad_var = data["svrg"]["grad_variance"]

# remove None values
passes_clean = []
grad_clean = []

for p, g in zip(passes, grad_var):
    if g is not None:
        passes_clean.append(p)
        grad_clean.append(g)

ax.semilogy(passes_clean, grad_clean, linewidth=2)

ax.set_title("SVRG Gradient Variance (Log Scale)")
ax.set_xlabel("Effective Passes")
ax.set_ylabel("Gradient Variance")
ax.grid(True)

# =========================
# 4. BEST TEST ERROR BAR CHART
# =========================

ax = axes[1, 1]

best_errors = []

for algo in algorithms:
    best_errors.append(min(data[algo]["test_error"]))

ax.bar(
    [a.upper() for a in algorithms],
    best_errors
)

ax.set_title("Best Test Error Comparison")
ax.set_ylabel("Best Test Error (%)")
ax.grid(True, axis="y")

# =========================
# FINAL LAYOUT
# =========================

plt.suptitle(
    f'Optimization Comparison on {data["dataset"].upper()}',
    fontsize=16
)

plt.tight_layout()
plt.show()
```
